```python
import pandas as pd
import os
from datetime import datetime

csv_path = r"C:\Users\Admin\Desktop\ETL_Extract_QueenEstherKibegi_668897\Shopping_Trends.csv"
state_file = "last_extraction_time.txt"

# Read the CSV file
df = pd.read_csv(csv_path)

# Show column names to identify the correct date column
print("Columns in CSV:", df.columns.tolist())

# Replace 'Order Date' below with the actual date column name from your CSV
date_column = "Order Date"  # <-- Change this if your column name is different

# Simulate a last extraction time
if os.path.exists(state_file):
    with open(state_file, "r") as f:
        last_extraction_time = f.read().strip()
else:
    last_extraction_time = ""  # Extract all if no state file

# Ensure the date column is in datetime format
df[date_column] = pd.to_datetime(df[date_column])

# Filter for new or updated records
if last_extraction_time:
    last_time = pd.to_datetime(last_extraction_time)
    new_rows = df[df[date_column] > last_time]
else:
    new_rows = df

# Print how many rows were extracted
print(f"Extracted {len(new_rows)} rows incrementally since last check.")

# Always update the timestamp file with the current timestamp (date and time)
current_time = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
with open(state_file, "w") as f:
    f.write(current_time)
```
